```python
import matplotlib.pyplot as plt
import csv

def read_celebrities_from_csv(file_path):
    with open(file_path, mode='r', encoding='utf-8') as file:
        return list(csv.reader(file))[1:]

def calculate_accuracy(file_path):
    data = read_celebrities_from_csv(file_path)
    total = 0
    correct = 0

    for row in data:
        if row[1].lower() != "unknown":
            total += 1
            if row[0].lower() == row[2].lower():
                correct += 1

    return (correct / total) * 100 if total > 0 else 0

def plot_comparison(accuracies, title, labels):
    colors = ['#1f77b4', '#ff7f0e'] * 2
    plt.figure(figsize=(10, 6))
    x = range(len(accuracies))
    plt.bar(x, accuracies, color=colors)
    plt.xticks(x, labels)
    plt.ylabel('Accuracy (%)')
    plt.title(title)
    plt.ylim(0, 100)
    plt.show()
```

```python
# Celebrities
accuracy_celeb_incontext_gpt4 = calculate_accuracy('data/celebrities_incontext_gpt-4.csv')
accuracy_celeb_outcontext_gpt4 = calculate_accuracy('data/celebrities_gpt-4.csv')
accuracy_celeb_incontext_gpt35 = calculate_accuracy('data/celebrities_incontext_gpt-3.5-turbo.csv')
accuracy_celeb_outcontext_gpt35 = calculate_accuracy('data/celebrities_gpt-3.5-turbo.csv')
plot_comparison([accuracy_celeb_incontext_gpt4, accuracy_celeb_outcontext_gpt4, 
                 accuracy_celeb_incontext_gpt35, accuracy_celeb_outcontext_gpt35], 
                "Celebrity Mother Reversal Accuracies", ["In-Context GPT-4", "Out-Context GPT-4", 
                                         "In-Context GPT-3.5", "Out-Context GPT-3.5"])

# Legislators
accuracy_legis_incontext_gpt4 = calculate_accuracy('data/legislators_incontext_gpt-4.csv')
accuracy_legis_outcontext_gpt4 = calculate_accuracy('data/legislators_gpt-4.csv')
accuracy_legis_incontext_gpt35 = calculate_accuracy('data/legislators_incontext_gpt-3.5-turbo.csv')
accuracy_legis_outcontext_gpt35 = calculate_accuracy('data/legislators_gpt-3.5-turbo.csv')
plot_comparison([accuracy_legis_incontext_gpt4, accuracy_legis_outcontext_gpt4, 
                 accuracy_legis_incontext_gpt35, accuracy_legis_outcontext_gpt35], 
                "Legislator Mother Reversal Accuracies", ["In-Context GPT-4", "Out-Context GPT-4", 
                                          "In-Context GPT-3.5", "Out-Context GPT-3.5"])

# Presidents
accuracy_pres_incontext_gpt4 = calculate_accuracy('data/presidents_spouses_incontext_gpt-4.csv')
accuracy_pres_outcontext_gpt4 = calculate_accuracy('data/presidents_spouses_gpt-4.csv')
accuracy_pres_incontext_gpt35 = calculate_accuracy('data/presidents_spouses_incontext_gpt-3.5-turbo.csv')
accuracy_pres_outcontext_gpt35 = calculate_accuracy('data/presidents_spouses_gpt-3.5-turbo.csv')
plot_comparison([accuracy_pres_incontext_gpt4, accuracy_pres_outcontext_gpt4, 
                 accuracy_pres_incontext_gpt35, accuracy_pres_outcontext_gpt35], 
                "Presidential Spouse Reversal Accuracies", ["In-Context GPT-4", "Out-Context GPT-4", 
                                                   "In-Context GPT-3.5", "Out-Context GPT-3.5"])
```
